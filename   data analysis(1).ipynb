```python
#importing libraries
import numpy as np #used for mathematical operation on arrays
import pandas as pd #analyze big data and make conclusions based on statistical theories
import matplotlib.pyplot as plt #Matplotlib is a python library used to create 2D graphs and plots by using python scripts.
#makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc
import seaborn as sns #a library for making statistical graphics in Python
```

```python
#reading data
df = pd.read_csv(r'Diwali Sales Data.csv', encoding='unicode_escape')
#we use unicode for conversion of non ascii values to repecctive asci value
#we use 'r' for avoiding errors of hyphins in memory location of file
```

```python
pd.isnull(df)
#shows false for not null values and true respectivly
```

```python
df.dropna(inplace=True)
#delete the null value and saving it !
```

```python
#change data type of a column
df['Amount']=df['Amount'].astype('int')
```

```python
#Exploratory Data Analysis
#gender
ax = sns.countplot(y='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
#containers stores the value of the graph
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
#Grouping by gender part of the code groups the DataFrame df by the 'Gender' column
#['Amount'].sum(): Within each group created by the 'Gender' column, this part calculates the sum of the 'Amount' column.
# this part of the code sorts the resulting DataFrame based on the 'Amount' column in descending order 
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Amount', y='Gender',data=sales_gen)
```

```python
#data analysis w.r.t GENDER
```

```python
ax = sns.countplot(y = 'Age Group', hue='Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#from above graph we can see that most of the buyers are of the age group between 26-35 yrs
```

```python
#States
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'State', y='Orders')
```

```python
#total numbers of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(15)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state,x='State',y='Orders')
```

```python
#total amount/sales from top 10 states
sales_statesales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state,x='State',y='Orders')
```

```python
#From above graphs we can see that unexpectidly orders are from uttar pradesh,
#Maharsahter and karnataka respectivly but total sales/amount is from up,karnataka and then MAharshtra
```

```python
#Martial Status
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state =df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

```python
#From above graphs we can see that most of the buyers are married(women) and they have high purcahsing power
```

```python
#Product Category
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
